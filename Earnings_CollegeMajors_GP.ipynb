## Visualizing Earnings Based on College Majors

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe(include=["number"])
```

```python
raw_data_count = recent_grads.shape
print(raw_data_count)

recent_grads = recent_grads.dropna()

cleaned_data_count = recent_grads.shape
print(cleaned_data_count)
```

### Using Scatter Plots to Explore Relationships 

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter")
```

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter")
```

```python
recent_grads.plot(x="Full_time", y="Median", kind="scatter")
```

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter")
```

```python
recent_grads.plot(x="Men", y="Median", kind="scatter")
```

```python
recent_grads.plot(x="Women", y="Median", kind="scatter")
```

### Observations
 - There does not seem to be a big difference in increaded major popularity and salary. 
 - There is no correlation between percentage of women in a major and median salary.
 - There does not seem to be a strong link between number of full-time mployees and median salary.

### Histograms to Explore Relations

```python
fig = plt.figure(figsize=(10,30))
ax1 = fig.add_subplot(8,1,1)
ax2 = fig.add_subplot(8,1,2)
ax3 = fig.add_subplot(8,1,3)
ax4 = fig.add_subplot(8,1,4)
ax5 = fig.add_subplot(8,1,5)
ax6 = fig.add_subplot(8,1,6)
ax7 = fig.add_subplot(8,1,7)
ax8 = fig.add_subplot(8,1,8)

cols = ["Sample_size","Median","Employed","Full_time","ShareWomen",
       "Unemployment_rate","Men","Women"]
axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8]

for i in range(8):
    axes[i].hist(recent_grads[cols[i]])
    axes[i].set_title(cols[i])
```

#### Observations
 - The most common median salary range is from $30,000 to $40,000
 - The percent of majors that are predominatly male versus perdominatly female are split about 50/50
