# Star Wars Survey

### Reading in the Data

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

### Cleaning the Data
 - Drop rows where respondent ID is null
 - Rename columns to be more discriptive
 - Convert yes/no values to Boolean

```python
star_wars = star_wars[star_wars["RespondentID"].notnull()]
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

cols = ["Have you seen any of the 6 films in the Star Wars franchise?", "Do you consider yourself to be a fan of the Star Wars film franchise?"]

for col in cols:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
star_wars.head()
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
names = {
    "Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
    "Unnamed: 4": "seen_2", 
    "Unnamed: 5": "seen_3", 
    "Unnamed: 6": "seen_4", 
    "Unnamed: 7": "seen_5", 
    "Unnamed: 8": "seen_6"
    }
```

```python
star_wars = star_wars.rename(columns = names)
```

```python
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
ranks = {
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        }
```

```python
star_wars = star_wars.rename(columns = ranks)
```

```python
star_wars.columns
```

### Compute average ranking for each movie
 - Respondents asked to rank each movie, with 1 being their favorite and 6 being their least favorite; a lower ranking is better

```python
%matplotlib inline
import matplotlib.pyplot as plt

blue = (0/255, 107/255, 164/255)

star_wars[star_wars.columns[9:15]].mean()

fig, ax = plt.subplots()
plt.bar(range(1,7), star_wars[star_wars.columns[9:15]].mean(),
        align= "center", color= blue)
plt.title("Ranking per Star Wars Movie")

for key,spine in ax.spines.items():
        spine.set_visible(False)
        ax.tick_params(bottom="off", top="off", right="off")
```

 - The data has been read in, cleaned, and columns have been given more descriptive names.
 - The mean of the ranking for each movie was computed- 1 was the highest score, and 6 was the lowest, so a lower ranking is better
 - From this, we can see the fifth movie is the favorite, followed by the fourth and sixth at an almost equal rating.
 - The prequels ranked far worse than the original triology.

### Compute how many people have seen each movie

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
orange = (255/255, 128/255, 14/255)

fig, ax = plt.subplots()
plt.bar(range(1,7), star_wars[star_wars.columns[3:9]].sum(),
       align= "center", color= orange)
plt.title("Number of Views per Star Wars Movie")

for key,spine in ax.spines.items():
        spine.set_visible(False)
        ax.tick_params(bottom="off", top="off", right="off")
```

 - The original triology was seen more than the prequels- this lends itself to the earlier rankings where the originals were more popular

### Star Wars preferences by gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
#determine average rankings for each movie by gender
men_means = males[males.columns[9:15]].mean()
women_means = females[females.columns[9:15]].mean()

#colors for use in bar graph
light_green = (188/255, 189/255, 34/255)
light_blue = (23/255, 190/255, 207/255)

ind = np.arange(6)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
men_bar = ax.bar(ind, men_means, width, color= light_green)
women_bar = ax.bar(ind + width, women_means, width, color= light_blue)

# add some text for labels, title and axes ticks
ax.set_ylabel('Avg. Ranking')
ax.set_title('Ranking by Gender')
ax.set_xticks(ind + width)
ax.set_xticklabels(('I', 'II', 'III', 'IV', 'V', 'VI'))

ax.legend((men_bar[0], women_bar[0]), ('Men', 'Women'))

for key,spine in ax.spines.items():
            spine.set_visible(False)
            ax.tick_params(bottom="off", 
                           top="off", 
                           right="off", 
                           left="off")
            
plt.show()
```

```python
#determine number of views for each movie by gender
men_views = males[males.columns[3:9]].sum()
women_views = females[females.columns[3:9]].sum()

fig, ax = plt.subplots()
men_bar = ax.bar(ind, men_views, width, color= light_green)
women_bar = ax.bar(ind + width, women_views, width, color= light_blue)

# add some text for labels, title and axes ticks
ax.set_ylabel('Number Views')
ax.set_title('Views by Gender')
ax.set_xticks(ind + width)
ax.set_xticklabels(('I', 'II', 'III', 'IV', 'V', 'VI'))

#Shrink current axis
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

#place legend to the right of axis 
ax.legend((men_bar[0], women_bar[0]), ('Men', 'Women'),
         loc='center left', bbox_to_anchor=(1, 0.5))

for key,spine in ax.spines.items():
            spine.set_visible(False)
            ax.tick_params(bottom="off", 
                           top="off", 
                           right="off", 
                           left="off")
            
plt.show()
```

 - More males watched the movies.
 - Females ranked the originals higher than males did.
